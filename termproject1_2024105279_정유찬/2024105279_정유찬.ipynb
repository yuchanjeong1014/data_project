## 4. 데이터 불러오기 및 가공하기

```python
import pandas as pd
```

```python
df0 = pd.read_csv("이혼건수_2015_2023.csv", encoding="euc_kr")
df1 = pd.read_csv("혼인건수_2015_2023.csv",encoding="euc-kr")
df2 = pd.read_csv("동거기간별_이혼수_2015_2023.csv", encoding="euc_kr")
df3 = pd.read_csv("미성년자녀수별_이혼수_2015_2023.csv", encoding="euc_kr")
df4 = pd.read_csv("직업별_이혼수_2015_2023.csv", encoding="euc_kr")
```

###  4-0. 이혼건수 자료 다듬기

```python
df0.drop(1,inplace=True)
df0.set_index('기본항목별',inplace=True)
df0 = df0.astype(int)
df0.reset_index('기본항목별',inplace=True)
df0
```

### 4-1. 가설 1 데이터 다듬고 그래프 출력 후 분석하기

#### 원본데이터 출력

```python
df1
```

#### 데이터 다듬기

```python
df1_1 = df0.loc[:,'2015'].div(df1.loc[0,'2015'])   #각 년도의 이혼 건수를 같은 해의 혼인 건수로 나눔
df1_2 = df0.loc[:,'2016'].div(df1.loc[0,'2016'])
df1_3 = df0.loc[:,'2017'].div(df1.loc[0,'2017'])
df1_4 = df0.loc[:,'2018'].div(df1.loc[0,'2018'])
df1_5 = df0.loc[:,'2019'].div(df1.loc[0,'2019'])
df1_6 = df0.loc[:,'2020'].div(df1.loc[0,'2020'])
df1_7 = df0.loc[:,'2021'].div(df1.loc[0,'2021'])
df1_8 = df0.loc[:,'2022'].div(df1.loc[0,'2022'])
df1_9 = df0.loc[:,'2023'].div(df1.loc[0,'2023'])
df1_f = pd.concat([df1_1,df1_2,df1_3,df1_4,df1_5,df1_6,df1_7,df1_8,df1_9],axis = 1)
df1_final = (df1_f*100).round(1)     # 소수 둘째자리에서 반올림(가독성을 위함)
df1_final = df1_final.transpose()     #그래프를 쉽게 그리고 가독성을 위해 행,열을 서로 전환
df1_final.columns = ['혼인 건수 대비 이혼 수치']   # 열 이름 재설정
```

#### 가설 1 최종 데이터 프레임 출력 및 설명

```python
df1_final
```

#### 위 데이터프레임은 각 년도별로  이혼건수를 같은 해의 혼인 건수로 나누고 100을 곱한 후 소수 첫째자리까지 나타낸 것이다. 각 수치는 편의상 이혼수치라고 하자. 이혼수치는 가독성을 위해 100을 곱한 값이므로 단위가 없다. 이혼수치가 높을 수록 혼인율 대비 이혼율이 높다는 것을 의미한다.

#### 가설 1 최종 데이터 프레임으로 그래프 그리기

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False
df1_final.plot.bar(width=0.7,rot=0,figsize=(7,4)).get_legend().remove()
plt.xlabel('년도',fontsize=12)
plt.ylabel('이혼 수치((이혼건수/혼인건구)*(100))',fontsize=12)
plt.title('년도에 따른 혼인 건수 대비 이혼 수치 그래프', fontsize=15,fontweight='bold')
plt.show()
```

#### 가설 1 그래프 분석하기
##### 위 그래프는 각 년도에 따른 혼인율 대비 이혼 수치를 나타낸 그래프이다. 그래프로 보아 2021년까지는 혼인 건수대비 이혼 건수가 증가하고 있지만 2022년부터는 다시 감소하고 있다. <span style ="color:red">2022, 2023년의 이혼 수치는 2015,2016년과 같은 옛날보다는 높지만 2021년에 비해 줄어들었다는 점에서 '혼인율 대비 이혼율이 점점 증가할 것이다'라는 가설 1은 틀렸다고 볼 수 있다.</span>

### 4-2. 가설 2 데이터 다듬고 그래프 출력 후 분석하기

#### 원본 데이터 출력

```python
df2   
```

#### 데이터 다듬기

```python
df2.drop(df2.index[9:],inplace=True)     #전국 데이터만 뽑기(각 도별 데이터 삭제)
df2.drop(0,axis=0,inplace=True)
df2.drop(columns='시도별',inplace=True)
df2.drop(5,axis=0,inplace=True)
df2_1=(df2.loc[1:,'2015'].div(df0.loc[0,'2015'])*(10**3)).round(0)   #전국데이터를 같은 해의 이혼 건수로 나누기
df2_2=(df2.loc[1:,'2016'].div(df0.loc[0,'2016'])*(10**3)).round(0)
df2_3=(df2.loc[1:,'2017'].div(df0.loc[0,'2017'])*(10**3)).round(0)
df2_4=(df2.loc[1:,'2018'].div(df0.loc[0,'2018'])*(10**3)).round(0)
df2_5=(df2.loc[1:,'2019'].div(df0.loc[0,'2019'])*(10**3)).round(0)
df2_6=(df2.loc[1:,'2020'].div(df0.loc[0,'2020'])*(10**3)).round(0)
df2_7=(df2.loc[1:,'2021'].div(df0.loc[0,'2021'])*(10**3)).round(0)
df2_8=(df2.loc[1:,'2022'].div(df0.loc[0,'2022'])*(10**3)).round(0)
df2_9=(df2.loc[1:,'2023'].div(df0.loc[0,'2023'])*(10**3)).round(0)
df2_f1=pd.concat([df2.loc[:,'동거기간별'],df2_1,df2_2,df2_3,df2_4,df2_5,df2_6,df2_7,df2_8,df2_9],axis=1) #추출한 각 해의 데이터 다시 합치기
df2_f1.set_index('동거기간별',inplace=True)      # 각 데이터의 숫자를 보기좋게 정수부분만 추출(소수점은 결과에 영향을 안 미칠 정도로 작음)
df2_f1 = df2_f1.astype(int)
df2_f1.reset_index('동거기간별',inplace=True)
df2_f2 = df2_f1.transpose()             # 보기 좀더 편하고 그래프를 쉽게 그릴 수 있게 행,열을 서로 전환
df2_final = df2_f2.loc['2015':,0:]
df2_final.columns=['0~4년','5~9년','10~14년','15~19년','20~24년','25년~29년','30년 이상' ]    #각 컬럼 이름 설정
```

#### 가설 2 최종 데이터프레임 출력 및 설명

```python
df2_final
```

 ##### 위의 행렬은 각 년도의 동거기간 별 이혼 건수를 같은 해의 총 혼인 건수로 나눈 수치의 1000을 곱한 값들이다. 이 값들을 편의상 이혼수치라고 하겠다. 이혼수치가 높을 수록 이혼을 하는 부부가 많다는 뜻이다.

#### 가설 2 최종 데이터 프레임으로 그래프 그리기

```python
df2_final.plot(linestyle = 'dashed',marker='o',linewidth=2,figsize=(9.5,6))
plt.legend(fontsize=8,loc='upper right', bbox_to_anchor=(1.145,1))
plt.xlabel('년도',fontsize=12)
plt.ylabel('이혼 수치((이혼건수/그 해의 총 이혼건수)*(10**4))',fontsize=12)
plt.title('년도에 따른 동거기간별 이혼수치', fontsize=15,fontweight='bold')
plt.show()
```

#### 가설 2 그래프 분석
##### 위 그래프는 년도에 따른 동거기간별 이혼수치 그래프이다. 이혼까지의 동거기간이 0-4년인  사람들은 가설과 달리 시간이 지날수록 점점 감소하는 경향을 보였고 5-9년, 10-14년, 15-19년인 사람들은 2021년까진 감소했다가 이후 조금씩 증가하는 경향을 보였다. 반면, 20-24년과 25-29년인 사람들은 2021년까지 증가했다가 다시 감소하는 경향을 보였다. 놀라운 건 30년 이상인데 2021년까지 다른 기간에 비해 유난히 크게 증가했다가 이후 감소하는 경향을 보였다.<br> 
##### 0-4년, 10-14년, 15-19년인 사람들이 매년 제일 높고 0-4년은 2023년을 제외하면 그중에서도 가장 높은 수치를 기록하고 있다. 하지만 이건 가설에서 처럼 이혼까지의 동거기간이 짧은 사람들이 늘어나고 있는것이 아닌 이혼하는 사람들 중엔 동거기간이 짧은 편에 속하는 사람들이 많다는 것을 의미 한다. <span style ="color:red">오히려 그래프 상으론 0-4년이 매년 감소하고 있고 동거기간이 길었던 사람들, 그중에서도 30년 이상이었던  사람들의 이혼수치가 증가하는 경향을 보이므로 이 가설 2는 틀렸다고 볼 수 있다.</span>
